## Exercise 1: Take several pictures of red, blue, and green items with your phone or other digital camera (or download some from the internet, if a camera isnâ€™t available).
### a. Load each image, and convert it to a tensor.
### b. For each image tensor, use the .mean() method to get a sense of how bright the image is.
### c. Take the mean of each channel of your images. Can you identify the red, green, and blue items from only the channel averages?

```python
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, threshold=50)
import imageio
import matplotlib.pyplot as plt

print("a.")
forestfire_arr = imageio.imread('photos/forestfire.jpg')
coral_arr = imageio.imread('photos/coral.jpg')
rosemary_arr = imageio.imread('photos/rosemary.jpg')

forestfire = torch.from_numpy(forestfire_arr).float()/255
coral = torch.from_numpy(coral_arr).float()/255
rosemary = torch.from_numpy(rosemary_arr).float()/255

print("forestfire.size() =", forestfire.size())
print("coral.size() =", coral.size())
print("rosemary.size() =", rosemary.size())
```

```python
print("b., c.")
forestfire_bw = torch.mean(forestfire, 2)
coral_bw = torch.mean(coral, 2)
rosemary_bw = torch.mean(rosemary, 2)

fig, axs = plt.subplots(3, 5, figsize=(20, 10))
axs[0, 0].imshow(forestfire)
axs[0, 1].imshow(forestfire[:, :, 0], cmap = "Reds")
axs[0, 2].imshow(forestfire[:, :, 1], cmap = "Greens")
axs[0, 3].imshow(forestfire[:, :, 2], cmap = "Blues")
axs[0, 4].imshow(forestfire_bw, cmap = "Greys")
axs[1, 0].imshow(coral)
axs[1, 1].imshow(coral[:, :, 0], cmap = "Reds")
axs[1, 2].imshow(coral[:, :, 1], cmap = "Greens")
axs[1, 3].imshow(coral[:, :, 2], cmap = "Blues")
axs[1, 4].imshow(coral_bw, cmap = "Greys")
axs[2, 0].imshow(rosemary)
axs[2, 1].imshow(rosemary[:, :, 0], cmap = "Reds")
axs[2, 2].imshow(rosemary[:, :, 1], cmap = "Greens")
axs[2, 3].imshow(rosemary[:, :, 2], cmap = "Blues")
axs[2, 4].imshow(rosemary_bw, cmap = "Greys")
```
